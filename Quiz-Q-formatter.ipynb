```python
def answers_listify(strings, right_or_wrong):
    out = ""
    if right_or_wrong == "right":
        a = "cor"
        b = "true"
        x = 'c'
    else:
        a = "wrong"
        b = "false"
        x = 'b'
        
    for n, s in enumerate(strings):
        out += f"ta_{a}[{n+1}]: [{x}{n+1}, {b}, \"{s}\"];\n"
    
    return out
```

```python
def feedback_listify(strings, right_or_wrong):
    out = ""
    if right_or_wrong == "right":
        a = "correct"
        b = "true" 
    else:
        a = "wrong"
        b = "false"
        
    for n, s in enumerate(strings):
        out += f"{a}_fd[{n+1}]: {s}\n"
    
    return out
```

```python
def push_listify(n, right_or_wrong):
    out = ""
    
    for i in range(n):
        if right_or_wrong == "right":
            out += f"if (not member (c{i+1}, stu_ans) and member(c{i+1}, ta_corset)) then push(correct_fd[{i+1}], specific_fb)\n"
        else:
            out += f"if member (b{i+1}, stu_ans) then push(wrong_fd[{i+1}], specific_fb)\n"
            
    return out
```

```python
def process_question_doc(question):
    lines = question.split("\n\n")
    lines.pop()
    lines.pop()

    count = 0

    # these boolean variables used to keep track of where to put feedback
    # if both false, last line was feedback
    last_line_right = False
    last_line_wrong = False

    right_ans = []
    wrong_ans = []

    right_feedback = []
    wrong_feedback = []

    for line in lines:
        current_line_feedback = False

        line = line.strip()
        feedback = ""

        # grab question number and question text
        if count == 0:
            question_no = line
            if "manual input" in question_no:
                print(f"Manual entry required for {question_no}")
                return None, None, None, None, None, None
        elif count == 1:
            question_text = line.replace("\"", "'")

        # otherwise process answers and feedback
        if count > 1:

            # check if current line is feedback for previous line
            if line[0] == "\"" and line[-1] == "\"":
                feedback = line
                feedback = "\"" + feedback[1:-1].replace("\"", "'") + "\""
                current_line_feedback = True

            # add appropriate feedback
            if last_line_right: right_feedback.append(feedback)
            elif last_line_wrong: wrong_feedback.append(feedback)

            if current_line_feedback:
                last_line_right = False
                last_line_wrong = False

            else:
                # otherwise, add line to relevant list
                if "*" in line:
                    right_ans.append(line.replace("*", "").replace("\"", "'"))
                    last_line_right = True
                    last_line_wrong = False

                elif len(line) >= 1: 
                    wrong_ans.append(line.replace("\"", "'"))
                    last_line_right = False
                    last_line_wrong = True

        count += 1

    # account for answers at the end without feedback not having blank feedback added
    while (len(right_feedback) < len(right_ans)):
        right_feedback.append("")
    while (len(wrong_feedback) < len(wrong_ans)):
        wrong_feedback.append("")

    return question_no, question_text, right_ans, wrong_ans, right_feedback, wrong_feedback
```

```python
# lesson template
def gen_question_text(question_name, question_text, n_option, n_corr_options, n_cor, correct_list, n_wrong, wrong_list,
                      wrong_feedback_list, corr_feedback_list, push_wrong_feedback, push_cor_feedback, scoring):
    
    temp = "&quot;335551550&quot;:6,&quot;335551620&quot;:6"
    start_chunk = f"""
      <question type="stack">
        <name>
          <text>{question_name}</text>
        </name>
        <questiontext format="html">
          <text><![CDATA[<p><span class="TextRun SCXW8825275 BCX0" lang="EN-AU" xml:lang="EN-AU" data-contrast="none">{question_text}</span><span class="EOP SCXW8825275 BCX0" data-ccp-props="{temp}"> </span></p>
    <p><span style="font-size: 0.875rem;">[[input:ans1]] [[validation:ans1]] [[feedback:prt1]]</span></p>]]></text>
        </questiontext>
        <generalfeedback format="html">
          <text></text>
        </generalfeedback>
        <defaultgrade>1.0000000</defaultgrade>
        <penalty>0.1000000</penalty>
        <hidden>0</hidden>
        <idnumber></idnumber>
        <stackversion>
          <text>2018080600</text>
        </stackversion>
        <questionvariables>
          <text><![CDATA[/*PART 1 - MCQ*/
    /*total number of statements showing to students and how many correct options are included*/
    n_option: {n_option}
    n_cor: {n_corr_options}

    /*create a list of correct options*/
    n_c: {n_cor}
    ta_cor: makelist([], n_c)
    {correct_list}
    ta_cor: rand_selection(ta_cor, n_cor);

    /*create a list of wrong options*/
    n_w: {n_wrong}
    ta_wrong: makelist([], n_w)
    {wrong_list}
    ta_wrong: rand_selection(ta_wrong, n_option - n_cor);

    /*combine two lists*/
    ta: append(ta_cor, ta_wrong);
    ta: random_permutation(ta);

    /*collect listed correct options*/
    ta_corans: maplist(first, sublist(ta, lambda([ex], second(ex))));
    ta_corset: setify(ta_corans);
    """
    middle_chunk = """
    /*wrong*/
    ta_wrongans: mcq_incorrect(ta);
    ta_wrongset: setify(ta_wrongans);]]></text>
        </questionvariables>
        <specificfeedback format="html">
          <text></text>
        </specificfeedback>
        <questionnote>
          <text>{@ta@}</text>
        </questionnote>
        <questionsimplify>1</questionsimplify>
        <assumepositive>0</assumepositive>
        <assumereal>0</assumereal>
        <prtcorrect format="html">
          <text><![CDATA[<p>Correct answer, well done.</p>]]></text>
        </prtcorrect>
        <prtpartiallycorrect format="html">
          <text><![CDATA[<p>Your answer is partially correct.</p>]]></text>
        </prtpartiallycorrect>
        <prtincorrect format="html">
          <text><![CDATA[<p>Incorrect answer.</p>]]></text>
        </prtincorrect>
        <multiplicationsign>dot</multiplicationsign>
        <sqrtsign>1</sqrtsign>
        <complexno>i</complexno>
        <inversetrig>cos-1</inversetrig>
        <logicsymbol>lang</logicsymbol>
        <matrixparens>[</matrixparens>
        <variantsselectionseed></variantsselectionseed>
        <input>
          <name>ans1</name>
          <type>checkbox</type>
          <tans>ta</tans>
          <boxsize>15</boxsize>
          <strictsyntax>1</strictsyntax>
          <insertstars>0</insertstars>
          <syntaxhint></syntaxhint>
          <syntaxattribute>0</syntaxattribute>
          <forbidwords></forbidwords>
          <allowwords></allowwords>
          <forbidfloat>0</forbidfloat>
          <requirelowestterms>0</requirelowestterms>
          <checkanswertype>0</checkanswertype>
          <mustverify>0</mustverify>
          <showvalidation>0</showvalidation>
          <options></options>
        </input>
        <prt>
          <name>prt1</name>
          <value>1.0000000</value>
          <autosimplify>1</autosimplify>
          <feedbackstyle>1</feedbackstyle>
          <feedbackvariables>
            <text><![CDATA[/*define student here*/
    stu_ans: ans1
    """
    feedback_chunk = f"""
    /*feedback for selected wrong answers*/
    wrong_fd: makelist([], {n_wrong})
    {wrong_feedback_list}

    /*feedback for not selected correct answers*/
    correct_fd: makelist([], {n_cor})
    {corr_feedback_list}

    /*define an empty list to collect feedback*/
    specific_fb: []; 

    /*test if the incorrect answer is selected by student*/
    /*if yes, add the feedback to the specific feedback list*/
    {push_wrong_feedback}

    /*test if the correct answer is not selected by student*/
    /*if yes, take no action, otherwise add the feedback to the specific feedback list*/
    {push_cor_feedback}

    /*Setify student's answer*/
    Sans_set: setify(ans1);

    /*The number of selected correct options*/
    n_selected_correct: length(intersect(Sans_set, ta_corset))

    /*The number of unselected incorrect options*/
    n_selected_wrong: length(intersect(Sans_set, ta_wrongset))
    n_unselected_wrong: n_option - n_cor - n_selected_wrong

    grade: (n_selected_correct + n_unselected_wrong)/n_option]]></text>
          </feedbackvariables>
          {scoring}
        </prt>
      </question>
    """
    
    return start_chunk + middle_chunk + feedback_chunk
```

```python
def gen_scoring(n_option):
    
    all_cor_node = f"""
    <node>
        <name>0</name>
        <answertest>AlgEquiv</answertest>
        <sans>(n_selected_correct + n_unselected_wrong)</sans>
        <tans>{n_option}</tans>
        <testoptions></testoptions>
        <quiet>0</quiet>
        <truescoremode>=</truescoremode>
        <truescore>1.0000000</truescore>
        <truepenalty></truepenalty>
        <truenextnode>-1</truenextnode>
        <trueanswernote>prt1-1-T</trueanswernote>
        <truefeedback format="html">
          <text><![CDATA[<p>You have correctly answered this part.</p>]]></text>
        </truefeedback>
        <falsescoremode>=</falsescoremode>
        <falsescore>0.0000000</falsescore>
        <falsepenalty></falsepenalty>
        <falsenextnode>1</falsenextnode>
        <falseanswernote>prt1-1-F</falseanswernote>
        <falsefeedback format="html">
          <text></text>
        </falsefeedback>
    </node>
    """
    
    letterise = ["zero", "one", "two", "three", "four", "five", "six", "seven", "eight", "nine", "ten"] 
    middle_chunk = ""
    
    # n = 1, 2, 3
    for n in range(1, n_option-1):
        middle_node = f"""
        <node>
            <name>{n}</name>
            <answertest>AlgEquiv</answertest>
            <sans>(n_selected_correct + n_unselected_wrong)</sans>
            <tans>{n_option-n}</tans>
            <testoptions></testoptions>
            <quiet>0</quiet>
            <truescoremode>=</truescoremode>
            <truescore>{1-(n/n_option)}</truescore>
            <truepenalty>0.0000000</truepenalty>
            <truenextnode>-1</truenextnode>
            <trueanswernote>prt1-{n+1}-T</trueanswernote>
            <truefeedback format="html">
              <text><![CDATA[<p>You have selected {"an" if n==1 else letterise[n]} incorrect answer{"s" if n>1 else ""} for this part.</p>
        """
        
        middle_node += """
        <p>{@specific_fb@}</p>]]></text>
            </truefeedback>
            <falsescoremode>-</falsescoremode>
            <falsescore>0.0000000</falsescore>
            <falsepenalty></falsepenalty>
        """
        
        middle_node += f"""
            <falsenextnode>{n+1}</falsenextnode>
            <falseanswernote>prt1-{n+1}-F</falseanswernote>
            <falsefeedback format="html">
              <text></text>
            </falsefeedback>
        </node>
        """
        
        middle_chunk += middle_node
        
    all_wrong_node = f"""
    <node>
        <name>{n_option-1}</name>
        <answertest>AlgEquiv</answertest>
        <sans>(n_selected_correct + n_unselected_wrong)</sans>
        <tans>1</tans>
        <testoptions></testoptions>
        <quiet>0</quiet>
        <truescoremode>=</truescoremode>
        <truescore>0.2000000</truescore>
        <truepenalty></truepenalty>
        <truenextnode>-1</truenextnode>
        <trueanswernote>prt1-{n_option}-T</trueanswernote>
        <truefeedback format="html">
          <text><![CDATA[<p>You have selected {letterise[n_option-1]} incorrect answers for this part.</p>
    """
    
    all_wrong_node += """
    <p>{@specific_fb@}</p>]]></text>
        </truefeedback>
        <falsescoremode>-</falsescoremode>
        <falsescore>0.0000000</falsescore>
        <falsepenalty></falsepenalty>
        <falsenextnode>-1</falsenextnode>
    """
    
    all_wrong_node += f"""
        <falseanswernote>prt1-{n_option}-F</falseanswernote>
        <falsefeedback format="html">
          <text><![CDATA[<p>You have not selected any correct answers for this part.</p>
    """
    
    all_wrong_node += """
    <p>{@specific_fb@}</p>]]></text>
        </falsefeedback>
      </node>
    """
    
    return all_cor_node + middle_chunk + all_wrong_node
```
